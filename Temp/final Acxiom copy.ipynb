```python
def encode(source, date_group):
    if source==1 and date_group=="21-22":
        return "DMC TS buyer"
    if source==2 and date_group=="21-22":
        return "Tour-no-buy no-VIP"
    if source==5 and date_group=="18-19":
        return "VIP Contract buyer (no-TS)"
    if source==4 and date_group=="21-22":
        return "VIP TS buyer"
    if source==3 and date_group=="18-19":
        return "VIP Contract buyer 18-19"
    if source==3 and date_group=="21-22":
        return "VIP Contract buyer"
    return "other"

full_df["source_name"] = full_df.apply(lambda x: encode(x["source"], x["date_group"]), axis=1)
```

```python
def region_encode(state_code):
    if state_code in ("CT", "ME", "MA", "NH", "RI", "VT"):
        return "New_England"
    elif state_code in ("DE", "DC", "MD", "NJ", "NY", "PA"):
        return "Mid_Atlantic"
    elif state_code in ("AL", "AR", "FL", "GA", "KY", "LA", "MS", "NC", "SC", "TN", "VA", "WV"):
        return "Southeast"
    elif state_code in ("IL", "MI", "OH", "WI", "IN"):
        return "Great_Lakes"
    elif state_code in ("IA", "KS", "MN", "MO", "NE", "ND", "SD"):
        return "Plains"
    elif state_code in ("CO", "ID", "MT", "UT", "WY"):
        return "Rocky_Mountain"
    elif state_code in ("AZ", "NV", "NM", "OK", "TX"):
        return "Southwest"
    elif state_code in ("AK", "CA", "HI", "OR", "WA"):
        return "Pacific"
    else:
        return "NA"


full_df["region"] = full_df.apply(lambda x: region_encode(x["state_code"]), axis=1)
```
