### Importing Vyper

```python
from vyper.user import Model
import pandas as pd
from sklearn.utils import shuffle

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 200)
```

### Read in Data

```python
%%time
df = pd.read_csv("../source_datasets/Epsilon Revenue Model Data.csv")
df = shuffle(df, random_state = 0)
df_oob = df.iloc[-3000:]
df = df.iloc[:-3000]
print(df_oob.shape, df.shape)
```

### Create Model

```python
Model?
```

```python
%%time
m = Model(data=df, 
          dependent_variable='Revenue',
          training_percentage=0.7, 
          na_drop_threshold=0.5,
          model_type='linear')
```

```python
%%time
#string_threshold: when data type is string and if no. of values increase this threshlod the variable is excluded
m.auto_define_variables(categorical_threshold = 30,
                        string_threshold = 30,
                        overwrite=True)
```

### Custom set variables

```python
m.set_variable_types({
   'EAI_PERFORMANCE_RANK': 'categorical', 
   'MIDDLE_INITITAL_1': 'exclude', #excluding a variable
   'ETHNIC_GROUP1': 'exclude',
   'ZIP': 'exclude',
   'PERSON_KEY_1': 'exclude',
   'VALUESCORE_CHANGE_DATE': 'exclude',
   'AGILITY_HH_KEY': 'exclude',
   'STREET_SUFFIX': 'exclude',
   'ZIP4': 'exclude',
   'FIPS_STATE_CODE' : 'exclude',
   'COUNTY_CODE': 'exclude', 
   'ID': 'exclude', 
   'FILE_CODE': 'exclude',
   'EMPTY_NESTER_TRIGGER': 'exclude',
   'CC_TRIGGER': 'exclude',
   'MERITSCORE': 'exclude', 
   'NUM_SOURC_VERIFY_HH': 'exclude',
   'VERIFICATION_DATE_HH': 'exclude',
   'top20': 'exclude',
   'INCOME_CHANGE_DATE': 'exclude',
   'VERIFICATION_DATE_PERSON1': 'exclude',
   'Revenue': 'target_variable'})
m.variables.show_types()
```

### Variable Profiling

```python
from vyper.user.explorer import DataProfiler
from openpyxl import Workbook
l = list(m.variables.get_active_variables())# to look at only active variables
l = l + ['Revenue']
var_prof_df = DataProfiler(df[l], dependent_variable='Revenue')
wb = Workbook()
var_prof_df.create_var_profiling_ws(wb=wb, sheet_name='profile', sort_by_variance = 'desc')
wb.save(r"variable profiles linear.xlsx")
```
