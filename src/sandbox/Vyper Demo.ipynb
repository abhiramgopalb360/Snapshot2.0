```python
from vyper.user import Model
import pandas as pd

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 200)
```

### Read in DataFrame

```python
wanted=['MT_ONLINE_DELIVERY_RESTAURANT_CUSTOMERS',
 'ACT_NUM_OFFLINE_PURCHASE_QUINT',
 'orderspercustomer',
 'MT_BUDGET_MEAL_PLANNERS',
 'MT_CORD_CUTTERS',
 'EAI_PERFORMANCE_RANK',
 'MT_EXPERIMENTAL_COOKS',
 'MT_MEAL_KIT_DELIVERY_CONSUMERS']
df=pd.read_csv('/home/charles/Documents/Epsilon Response Model Data.csv')
df = df[wanted + ['y']]  
```

### Create Model

```python
Model?
```
